# Linear regression with gradient descent numpy implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
file_path = "Student_Performance.csv"

df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "nikhil7280/student-performance-multiple-linear-regression",
  file_path
)
```

```python
df.head()
```

```python
df_transformed = df.copy()
```

```python
# Standardisation of features

numeric_cols = ['Hours Studied', 'Previous Scores', 'Sleep Hours', 'Sample Question Papers Practiced']
for col in numeric_cols:
  mean = np.mean(df[col])
  std_dev = np.std(df[col])
  df_transformed[col] = (df[col] - mean) / std_dev
```

```python
df_transformed.loc[df['Extracurricular Activities'] == 'Yes', 'Extracurricular Activities'] = 1
df_transformed.loc[df['Extracurricular Activities'] == 'No', 'Extracurricular Activities'] = 0
df_transformed.head()
```

```python
feature_cols = ['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']
X = df_transformed[feature_cols].values
y = df_transformed['Performance Index'].values #.reshape(-1, 1)
```

```python
X.shape[0]
```

```python
def SSR(y_real, y_pred):
  residuals = y_real - y_pred
  squared_res = np.square(residuals)
  sum = np.sum(squared_res)
  return sum
```

```python
def MSE(SSR, num_of_obs):
  return SSR / (num_of_obs*2)
```

```python
def R2(y_real, y_pred):
  mean_val = np.mean(y_real)
  mean_ssr = SSR(y_real, mean_val)
  pred_ssr = SSR(y_real, y_pred)
  return (mean_ssr - pred_ssr) / mean_ssr
```

```python
def predict(X, w, b):
  n = X.shape[0] # number of training examples
  y_pred = np.zeros(n) # create an 'empty' vector of predictions for the training examples
  for i in range(n):
    y_pred[i] = np.dot(X[i], w) + b # prediction for training example := dot product of X row and w vector
  return y_pred
```

```python
def gradient_descent(X, y, epochs=100, lr=0.3, show_descent=True):
  n = X.shape[0] # number of training examples
  m = X.shape[1] # number of features

  # initialise w vector and b to 0
  w = np.zeros(m)
  b = 0

  # w1_step = -1
  # b_step = -1

  cost_history = [] # keep track of cost as it is reducing

  for i in range(epochs): # perform 100 epochs
    y_pred = predict(X, w, b) # calculate vector of predicted y with current w and b
    residual = y_pred - y # vector of residuals

    # calculate derivatives
    dj_dw = np.zeros(m)
    for j in range(m):
      dj_dw[j] = np.dot(np.transpose(X)[j], residual) / n
    dj_db = np.sum(residual) / n

    # update parameters
    w -= dj_dw * lr
    b -= dj_db * lr

    # add cost to cost_history
    cost_history.append(MSE(SSR(y, y_pred), n))

    if show_descent:
      if i % 9 == 0: # every ten iterations
        print(f'Iteration: {i+1} – Cost: {cost_history[i]:2f} – ',
              f'dJ/dw1: {dj_dw[0]:2f} – dJ/db = {dj_db:2f} – ',
              f'w1: {w[0]:2f} – b: {b:2f}')

  return w, b, cost_history
```

```python
w, b, cost_history = gradient_descent(X, y)
```

```python
plt.plot(np.arange(100), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost minimisation with gradient descent')
plt.show()
```

```python
y_pred = predict(X, w, b)
plt.scatter(df['Previous Scores'], y)
plt.scatter(df['Previous Scores'], y_pred, color='red')
plt.title('Visualising actual vs predicted scores (predicted=red)')
plt.xlabel('Previous scores')
plt.ylabel('Performance index')
plt.show()
```

# K-Fold Cross Validation

```python
def k_fold_cross_val(k, df):
  # randomize the order of the data
  data = df_transformed.to_numpy()
  np.random.shuffle(data)

  # split the data into k equal groups
  k_folds = np.array_split(data, k) # returns list of np arrays

  # iteratively select k-1 groups to be the training data and the one group
  # that is left to be the validation data
  mse_history = []
  r2_history = []

  for fold in range(len(k_folds)):
    X_val = k_folds[fold][:, :5]
    y_val = k_folds[fold][:, 5]

    parts = k_folds[:fold] + k_folds[fold+1:]
    X_train = np.vstack(parts)[:, :5]
    y_train = np.vstack(parts)[:, 5]

    # train the model using the training data
    w, b, cost_history = gradient_descent(X_train, y_train, show_descent=False)

    # evaluate using the validation set
    y_pred = predict(X_val, w, b)
    mse = MSE(SSR(y_val, y_pred), np.size(y_val))
    r2 = R2(y_val, y_pred)

    # keep track of the entire history of evaluations for the model
    mse_history.append(mse)
    r2_history.append(r2)

  # return the evaluation history
  return mse_history, r2_history

mse_history, r2_history = k_fold_cross_val(10, df)
```

```python
print(f'mean MSE from CV: {np.mean(mse_history)}')
print(f'mean R2 from CV: {np.mean(r2_history)}')
```

```python
plt.scatter(np.arange(len(mse_history)), mse_history)
plt.title('MSE changing through cross-validation')
plt.xlabel('Folds')
plt.ylabel('MSE')
plt.show()
```

```python
plt.scatter(np.arange(len(r2_history)), r2_history)
plt.title('R2 changing through cross-validation')
plt.xlabel('Folds')
plt.ylabel('R2')
plt.show()
```
